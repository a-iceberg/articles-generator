# Пример обращения к апи

```python
import requests

files = {"file": ("iceberg.csv", open("iceberg.csv","rb"), "text/csv")}

data = {"groups_start": "0", "groups_end": "1", "save_html": "false", "keep_server_copy": "false"}

with requests.post("http://62.197.49.99:8001/articles_generator_upload",
                   files=files, data=data, stream=True, timeout=3600) as r:
    r.raise_for_status()

    groups = r.headers.get("X-Groups-Processed", "?")
    total  = r.headers.get("X-Total-Cost", "?")

    with open("articles.csv", "wb") as f:
        for chunk in r.iter_content(1<<14):
            if chunk:
                f.write(chunk)

print(f"ok. Обработано статей: {groups}. ИТОГОВАЯ сумма: ${total}")
```
